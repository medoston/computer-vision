```python
# ploting multible graphs
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.cos(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.plot(t,y1,"r", t,y2,"g", t,y3,"ys")
```

```python
#sup plot
# ploting multible graphs
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.cos(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.subplot(211)
plt.plot(t,y1,"r")
plt.subplot(212)
plt.plot(t,y2,"g")
plt.subplot(212)
plt.plot(t,y3,"ys")
```

```python
#sup plot
# ploting multible graphs
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.cos(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.subplot(131)
plt.xlabel("sin(x")
plt.plot(t,y1,"r")
plt.subplot(132)
plt.xlabel("cos(x)")
plt.plot(t,y2,"g")
plt.subplot(133)
plt.xlabel("tan(x)")
plt.plot(t,y3,"ys")
```

```python
dx = 0.01
dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)
def sum (x,y):
  return (1-y**5 + x**5)* np.exp(-x**2 - y**2)

C = plt.contour (X,Y,sum(X,Y),8,colors = 'lightPink')
plt.contourf(X,Y,sum(X,Y),8)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
dx = 0.01
dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)
def sum (x,y):
  return (1-y**5 + x**5)* np.exp(-x**2 - y**2)

# C = plt.contour (X,Y,sum(X,Y),8,colors = 'lightPink')
# plt.contourf(X,Y,sum(X,Y),8)
ax.plot_surface(X,Y,sum(X,Y), rstride = 1, cstride = 1)
```
